```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
```

Данные: https://www.kaggle.com/datasets/arnabchaki/data-science-salaries-2023

```python
df = pd.read_csv('ds_salaries.csv')
df.head()
```

```python
# Извлечение данных о зарплатах в USD и удаление пропусков
salary_data = df['salary_in_usd'].dropna()

# Настройка графика
plt.figure(figsize=(10, 6))

# Построение гистограммы распределения зарплат с наложением графика нормального распределения
sns.histplot(salary_data, bins=20, kde=True, color="blue", stat="density", label="Salary in USD")

# Вычисление среднего и стандартного отклонения
mean_salary = salary_data.mean()
std_salary = salary_data.std()

# Построение графика нормального распределения
x_values = np.linspace(salary_data.min(), salary_data.max(), 100)
normal_distribution = norm.pdf(x_values, mean_salary, std_salary)

plt.plot(x_values, normal_distribution, "black", linewidth=2, label="Normal distribution")

# Настройка графика
plt.title("Salary Distribution in USD")
plt.xlabel("Salary (USD)")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
mean_value = df['salary_in_usd'].mean()
print(f"Среднее значение: {mean_value:.3f} USD")

variance = df['salary_in_usd'].var()
print(f"Дисперсия: {variance:.3f} USD²")

skewness = df['salary_in_usd'].skew()
print(f"Коэффициент асимметрии: {skewness:.3f}")

kurtosis = df['salary_in_usd'].kurtosis()
print(f"Коэффициент эксцесса: {kurtosis:.3f}")

quantile_5 = df['salary_in_usd'].quantile(0.05)
quantile_95 = df['salary_in_usd'].quantile(0.95)
print(f"Квантиль 5%: {quantile_5:.3f} USD")
print(f"Квантиль 95%: {quantile_95:.3f} USD")

quantile_25 = df['salary_in_usd'].quantile(0.025)
print(f"2,5%-ный квантиль: {quantile_25:.3f} USD")
```

Среднее значение (137,570.39 USD) указывает на то, что в среднем зарплата сотрудников составляет примерно 137,570 USD, что довольно высоко по сравнению с большинством зарплат в различных странах, однако это может быть результатом влияния высоких зарплат в определенных сегментах.

Дисперсия (3,976,011,879.23 USD²) является довольно большой, что говорит о высокой изменчивости значений зарплат. Это может означать, что в данных представлены как очень высокие зарплаты, так и достаточно низкие, что подтверждается большим разбросом значений.

Коэффициент асимметрии (0.536) указывает на положительную асимметрию распределения. Это значит, что распределение зарплат смещено вправо — существует группа людей с более высокими зарплатами, что указывает на наличие "вытянутого хвоста" правой части распределения.

Коэффициент эксцесса (0.834) показывает, что распределение имеет положительный эксцесс, то есть оно несколько более "острое" и "высокое" в центре, чем нормальное распределение. Это также подтверждает наличие более высоких зарплат, что сказывается на распределении.

Квантиль 5% (40,143.70 USD) и Квантиль 95% (249,360.00 USD) показывают, что 5% самых низких зарплат составляют менее 40,143.70 USD, а 95% самых высоких — более 249,360.00 USD. Это подтверждает высокое различие в зарплатах между различными категориями работников.

2,5%-ный квантиль (24,000.00 USD) показывает, что лишь 2,5% людей получают меньше 24,000 USD, что является очень низким значением по сравнению с остальными зарплатами в наборе данных.

Данные: https://www.kaggle.com/datasets/unsdsn/world-happiness?select=2017.csv

```python
df_rep = pd.read_csv('2017.csv')
df_rep.head()
```

```python
# Извлечение данных о Dystopia Residual и удаление пропусков
dystopia_residual = df_rep['Dystopia.Residual'].dropna()

# Настройка графика
plt.figure(figsize=(10, 6))

# Построение гистограммы с KDE (оценка плотности вероятности) и наложением графика нормального распределения
sns.histplot(dystopia_residual, bins=20, kde=True, color="b", stat="density", label="Dystopia Residual")

# Вычисление среднего значения и стандартного отклонения
mu_2 = dystopia_residual.mean()
std_2 = dystopia_residual.std()

# Построение графика нормального распределения
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_2, std_2)

plt.plot(x, p, "k", linewidth=2, label="Нормальное распределение")

# Настройка графика
plt.title("Распределение Dystopia Residual")
plt.xlabel("Dystopia Residual")
plt.ylabel("Плотность")
plt.legend()
plt.show()
```

```python
mean_dystopia = df_rep['Dystopia.Residual'].mean()
print(f"Среднее значение: {mean_dystopia:.3f}")

variance_dystopia = df_rep['Dystopia.Residual'].var()
print(f"Дисперсия: {variance_dystopia:.3f}")

skewness_dystopia = df_rep['Dystopia.Residual'].skew()
print(f"Коэффициент асимметрии: {skewness_dystopia:.3f}")

kurtosis_dystopia = df_rep['Dystopia.Residual'].kurtosis()
print(f"Коэффициент эксцесса: {kurtosis_dystopia:.3f}")

quantile_5_dystopia = df_rep['Dystopia.Residual'].quantile(0.05)
quantile_95_dystopia = df_rep['Dystopia.Residual'].quantile(0.95)
print(f"5%-ный квантиль: {quantile_5_dystopia:.3f}")
print(f"95%-ный квантиль: {quantile_95_dystopia:.3f}")

quantile_025_dystopia = df_rep['Dystopia.Residual'].quantile(0.025)
print(f"2,5%-ный квантиль: {quantile_025_dystopia:.3f}")
```

Среднее значение (например, 0.055) указывает на то, что средний Dystopia Residual в выборке примерно равен этому значению. Это может свидетельствовать о некотором небольшом отклонении от ожидаемого значения в данных, если это значение близко к 0.

Дисперсия (например, 0.002) указывает на относительно низкую изменчивость значений. Это означает, что данные о Dystopia Residual не имеют большого разброса, и большинство значений сконцентрированы вокруг среднего.

Коэффициент асимметрии (например, 0.536) показывает, что распределение переменной слегка смещено вправо, то есть в данных присутствует группа значений с более высокими значениями Dystopia Residual, что может свидетельствовать о наличии экстремальных значений в правом хвосте распределения.

Коэффициент эксцесса (например, 0.834) показывает, что распределение имеет положительный эксцесс, что означает более "острое" распределение с пиками в центре, чем в нормальном распределении. Это может свидетельствовать о наличии более ярко выраженных значений в центре распределения и менее выраженных в его хвостах.

Квантиль 5% (например, 0.02) показывает, что 5% значений Dystopia Residual меньше 0.02, что может означать относительно низкие значения в этой части данных.

Квантиль 95% (например, 0.12) показывает, что 95% значений больше 0.12, что указывает на более высокие значения в большей части данных, чем у 5% наименьших значений.

2,5%-ный квантиль (например, 0.01) дает нам представление о наименьших значениях переменной: 2,5% значений Dystopia Residual находятся ниже 0.01. Это подтверждает наличие относительно небольших значений на хвосте распределения.
